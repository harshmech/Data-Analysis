```python
#Importing the important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('ggplot') #optional: for ggplot-like style
print('Matplotlib version:', mpl.__version__)
```

```python
#Fetching the DATA

df_can = pd.read_csv(r'D:\Data Analysis\Data Visualtion\Canada_w2.csv')
```

```python
# Creating a list of years from 1980-2013 as this will come handy while plotting 

years = list(map(str, range(1980,2013)))
years
```

```python
# Creating a STAKED LINE PLOT OR AREA PLOT

df_can.sort_values(['Total'], ascending = False, axis = 0, inplace=True)
#get the top 5 entries
df_top5 = df_can.head()
#transpose of the dataframe
df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
# area plots are stacked by default i.e each column must be pos. or neg. (NaN values are defaulted 
# to zero) for UNSTACKED PLOT  set parameter STACKED to FALSE
```

```python
#changinig the index values of df_top5 to type INT for plotting
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',
             stacked = False,
             figsize=(20,10))

plt.title('Immigration Trend of Top 5 Countries')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')

plt.show()
```

```python
# Difference between a unstacked and a stacked area plot
df_top5.plot(kind='area', alpha = 0.25, stacked = True, figsize=(20,10))



plt.title('Immigration trend for top 5 Countries')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')
```

```python
# example of Artist layer(Object Oriented Layer)

ax = df_top5.plot(kind='area', alpha = 0.75, figsize=(20,10))
ax.set_title('Immigration Trend of top 5 Countries')
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Years')
```

```python
df_bot5= df_can.tail()
df_bot5 = df_bot5[years].transpose()
df_bot5.index = df_bot5.index.map(int) # chag. the index values of df_bot5 to type integer 
                                     #for PLOTTING
```

```python
df_bot5.plot(kind='area', alpha = 0.7, stacked =  False, figsize=(20,10))
plt.title('Immigartion trend of 5 Countries with least contribution to Immigration')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
```

```python
hh = df_bot5.plot(kind='area', alpha=0.55, stacked = False, figsize = (30,10)) 
hh.set_title('Immigartion trend of 5 Countries with least contribution to Immigration')
hh.set_xlabel('Years')
hh.set_ylabel('Number of Immigrants')
```

```python
#HISTOGRAMS - way of representing the frequency distribution of numeric dataset.
# partitions the x-axis into bins, assigns each data point in our dataset to a bin
#  y-axis is the frequency or the number of data points in each bin.
```

```python
# data split -  Numpy's histrogram method to get the bin ranges and frequency counts 
```

```python
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
print(count)
print(bin_edges)
```

```python
#the default number of bins is 10 in the histogram.
#plotting the histogrm
df_can['2013'].plot(kind='hist', figsize=(8,5))
plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.xlabel('Number of Immigrants')
plt.ylabel('Number of Countries')
plt.show()
```

```python
#matching the x-axis labels with the bin size
count, bin_size = np.histogram(df_can['2013'])
df_can['2013'].plot(kind='hist', figsize=(8,5), xticks=bin_size)
plt.title('History of Immigration from 195 Countries')
plt.xlabel('Number of Immigrants')
plt.ylabel('Number of Countries')
plt.show()
```
