```python
#importing the libraries
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') #for ggplot-like style
print('Matplotlib version: ', mpl.__version__)
```

```python
df_can = pd.read_csv(r'D:\Data Analysis\Data Visualtion\Canada_w2.csv')
```

```python
df_can.head()
```

```python
#dimensions of the dataframe(no. of entries)
print(df_can.shape)
```

```python
#setting the countries as the  index, so as to quickly look up countries using the .loc method
df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
#seeing the change in data dimensions
print('data dimensions:', df_can.shape)
```

```python
#creating a list of years from 1980-2013 which will come in handy while plotting
years = list(map(str, range(1980,2014)))
years
```

```python
# 1. AREA PLOT
```

```python
#sorting the values in descending order
df_can.sort_values(['Total'], ascending=False, axis = 0, inplace=True)
df_top5 = df_can.head()
#transpose the dataframe
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
#changing the index values of df_top5 to type integar for plotting
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area', stacked=False, figsize=(10,10), alpha = 0.6)
#default alpha value is 0.5, alpha is the transparency index
plt.title('Immigration trend of top 5 countries')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')
plt.show()
```

```python
#PLOTTING USING THE SCRIPTING LAYER (MATPLOTLIB ASPLT) i.e. PROCEDURAL METHOD
```

```python
df_top5.plot(kind='area', figsize=(20,10), alpha=0.45)
plt.title('Immigration')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.show()
```

```python
#PLOTTING USING THE ARTISTIC LAYER(OBJ. ORIENTED METHOD)
ax = df_top5.plot(kind='area', figsize=(20,10), alpha=0.55)
ax.set_title('Immigration')
ax.set_xlabel('years')
ax.set_ylabel('Number of immigrans')
```
